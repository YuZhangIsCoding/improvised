# Generating wordclouds using python library WordCloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

## Simple case

Using simple setups

```python
# Read in text
text = open('loseyourself.txt').read()
```

```python
# Tune the max font size to 60, and background color to white
wc = WordCloud(max_font_size = 80, background_color= 'white', width = 800, height = 300).generate(text)

fig = plt.figure(figsize = (8, 3), dpi = 300)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.tight_layout()
```

## Simple mask
### Round shape

```python
x, y = np.ogrid[:500, :500]

mask = (x - 250) ** 2 + (y - 250) ** 2 > 230 ** 2
mask = 255 * mask.astype(int)

wc = WordCloud(background_color="white", mask=mask).generate(text)
fig = plt.figure(figsize = (3, 3), dpi = 300)
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

## Mask from image

```python
mymask = np.array(Image.open('eminem.jpg').convert('L'))

threshold = 180
mymask = (mymask > threshold)*255

plt.imshow(mymask, cmap='gray')
```

```python
wc = WordCloud(background_color="white", mask=mymask, max_font_size = 100,
              max_words=2000).generate(text)
fig = plt.figure(figsize = (3, 3), dpi = 300)
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```
